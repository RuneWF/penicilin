```python
x_tick_labels = [
    "Ecosystem\nDamage",
    "Human\nHealth\nDamage",
    "Natural\nResources\nScarcity"
    ]
```

```python
init = m.main()

impact_categories = init.lcia_impact_method()
plot_x_axis_all = [0] * len(impact_categories)
for i in range(len(plot_x_axis_all)):
    plot_x_axis_all[i] = impact_categories[i][2]

# Extract the midpoint categories from the plot x-axis
ic_mid = plot_x_axis_all[:-3]

plot_x_axis_mid = []
# Process each midpoint category to create a shortened version for the plot x-axis
for ic in ic_mid:
    string = re.findall(r'\((.*?)\)', ic)
    if 'ODPinfinite' in string[0]:
        string[0] = 'ODP'
    elif '1000' in string[0]:
        string[0] = 'GWP'
    plot_x_axis_mid.append(string[0])
```

```python
dct = {}
for col in results_df_mid.columns:
    df_temp = pd.read_excel(path, sheet_name="mid_to_end", index_col=0)
    df_temp.columns = results_df.columns[-3:]
    for idx, row in results_df_mid.iterrows():
        df_temp.loc[str(idx)] *= row[col]
    
    for tcol in df_temp.columns:
        df_temp[tcol] /= df_temp[tcol].sum()

    dct[col] = df_temp


for key, df in dct.items():
    idx_new = []
    for txt in plot_x_axis_mid:
        for idx in df.index:
            if txt in idx:
                idx_new.append(idx)
    df = df.reindex(idx_new)
    df.index = plot_x_axis_mid
    df.columns = x_tick_labels
    dct[key] = df
# idx_new
dct["Penicillin G, defined system"]
```

```python
def mid_to_end_contribution_title_text(act, ax):
    if "V" in str(act):
        return ax.set_title(f"Mid- to Endpoint contribution - Oral")
    else:
        return ax.set_title(f"Mid- to Endpoint contribution - IV")
```

```python
import matplotlib.pyplot as plt

width_in, height_in, dpi = init.plot_dimensions()
fig, axes = plt.subplots(1,2, figsize=(width_in, height_in), dpi=dpi)

plt.subplots_adjust(wspace=0.4)


for a, (act, df) in enumerate(dct.items()):

    ax = axes[a]

    df = df.T
    bars = df.plot(
        kind='bar',
        stacked=True,
        ax=ax,
        cmap="coolwarm",
        zorder=2
    )
    hatch_styles = ['', '\\o', 'xx', 'O', '++', "oo"]
    # Apply different hatch styles to each layer (row)
    for i, patch_list in enumerate(bars.containers):
        hatch = hatch_styles[i % len(hatch_styles)]
        for patch in patch_list:
            patch.set_hatch(hatch)

    mid_to_end_contribution_title_text(act, ax)
    ax.set_xticklabels(x_tick_labels, rotation=0)
    ax.grid(axis='y', linestyle='--', alpha=0.7, zorder=-0)
    ax.set_ylabel("Share of the impact")
    y_ticks = plt.gca().get_yticks()
    ax.set_yticks(y_ticks)
    ax.set_yticklabels(['{:.0f}%'.format(y * 100) for y in y_ticks])
    ax.set_ylim(0,1.01)
 
    if a == 1:
        leg_color, _ = fig.gca().get_legend_handles_labels()
        leg_txt = list(df.T.index)
        leg_txt = leg_txt[::-1]
        leg_color = leg_color[::-1]
        ax.legend(
            leg_color,
            leg_txt,
            loc='upper left',
            bbox_to_anchor=(0.995, 1.02),
            ncol=1,
            fontsize=10,
            frameon=False
            )
    else:
        ax.get_legend().remove()
plt.show()
```
